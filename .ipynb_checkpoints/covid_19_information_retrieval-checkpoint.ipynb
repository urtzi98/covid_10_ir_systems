# COVID19 Information Retival System


In this work, we want to build and evaluate a custom Informatin Retrival System. For that, it will use the TREC-COVID19 corpus to build an IR engine targeted at retrieving documents from such corpus which are relevant to different user queries related to SARS-COV-2 and the COVID-19 disease. The next resources are needed for the IR system to work:


1) **The TREC-COVID19 corpus:** It can be downloaded from this link: https://ai2-semanticscholar-cord-19.s3-us-west-2.amazonaws.com/historical_releases/cord-19_2020-07-16.tar.gz

For the project, the used corpus is located in **pdf_json** folder, inside **document_parses.tar.gz**

2) **Test set of 50 topics:** A file containing the set of queries which will be used to assess the performance of the information retrieval engine. There are 50 queries (topics in the TREC jargon) which will be used for the evaluation. It can be downloaded here: https://ir.nist.gov/covidSubmit/data/topics-rnd5.xml

3) **Some Relevance Judgements:** The relevance judgements are needed to turn a set of documents and topics into a test collection. Each line in this file provides a relevance score (either binary or non-binary) for a document given a topic (query). The relevance scores provided in this sample corpus are discrete {0, 1 and 2}, the lowest score meaning that the document is totally irrelevant w.r.t. the query and the highest entailing that the document is totally relevant. For this exercise we will use binary relevance judgements, meaning that documents with a nonzero score are relevant to the topic or irrelevant otherwise. Link to download: https://ir.nist.gov/covidSubmit/data/qrels-covid_d5_j0.5-5.txt

### Generate the coprus

Before implementing any model, we are going to generate the corpus. To do so, we are going to read the *metadata.csv* of the document database, preprocess those documents and save that preprocessed documents in the */data* folder. This is done to avoid loading all the corpus at once and also avoid memory problems.

```python
# Create directories for saving the preprocessed data
if not os.path.isdir("./data"):
    os.makedirs('./data')
if not os.path.isdir("./data/corpus"):
    os.makedirs('./data/corpus')
```

```python
def preprocess_text(text):
    stopset = set(stopwords.words('english'))
    stemmer = PorterStemmer()
    tokens = wordpunct_tokenize(text)
    clean = [token.lower() for token in tokens if token.lower() not in stopset and len(token) > 2]
    final = [stemmer.stem(word) for word in clean]
    return final


def process_corpus(path_pdf, path_to_save):
    path_pdfs = os.listdir(path_pdf)
    for pdf in path_pdfs:
        pdf = "../cord-19_2020-07-16/2020-07-16/document_parses/pdf_json/"+pdf
        with open(pdf) as json_file:
            data = json.load(json_file)
            if data['abstract'] == []:
                preprocessed_doc = {'sha': data['paper_id'],
                                    'title': preprocess_text(data['metadata']['title']),
                                    'abstract': [],
                                    'body_text': preprocess_text(data['body_text'][0]['text'])
                                    }
            else:
                preprocessed_doc = {
                                    'sha': data['paper_id'],
                                    'title': preprocess_text(data['metadata']['title']),
                                    'abstract': preprocess_text(data['abstract'][0]['text']),
                                    'body_text': preprocess_text(data['body_text'][0]['text'])
                                    }
            json.dump(preprocessed_doc, open(path_to_save+data['paper_id']+'.json', 'w+'))
```

```python
# Execute it only once
corpus = process_corpus("../cord-19_2020-07-16/2020-07-16/document_parses/pdf_json/", "./data/corpus/")
```

Also, as in the relevance jugements the documents are identifyed by their id, and in the corpus the files containing the information are saved with their SHA key, is necessary to have a file that relationates those two names, we are going to generate a file in which we are saving that information

```python
def create_equivalence_json(path_csv, path_pdfs):
    sha = [pdf.split(".")[0] for pdf in path_pdfs]
    id_uid = dict()
    with open(path_csv) as f_in:
        reader = csv.DictReader(f_in)
        for row in reader:
            if row['sha'] in sha:
                cord_uid = row['cord_uid']
                id_uid[cord_uid] = row['sha']
    json.dump(id_uid, open('./data/'+'cord_uid_sha_equivalences'+'.json', 'w+'))
```

### Generate a document with  the querys

As done before with the coprus, we are going to preprocess the querys and seave them into a document in the */data* directory

```python
def process_querys(path, path_to_save):
    topics = {}
    root = ET.parse(path).getroot()

    for topic in root.findall("topic"):
        topic_number = topic.attrib["number"]
        topics[topic_number] = {}
        for query in topic.findall("query"):
            topics[topic_number] = preprocess_text(query.text)
    json.dump(topics, open(path_to_save+"processed_querys.json", 'w+'))
```

### Generate test set from relevant judgements

Here we are going to load the relevant judgemetns and preprocess it. This means, we are goint to take into account the columns of: *topic_id*, *cord_uid* and *relevance_score*. The relevance score in the given file is discrete, {0, 1 or 2}, where 0 means totally irrelevant, 1 more or less relevant and 2 totally relevant. In our case, we are goint to make it binary, meaning '0' irrelevant and '1' relevant documents.

```python
def get_test_from_judgemetns(path):
    relevance_data = pd.read_csv(path, sep=" ", header=None)
    relevance_data.columns = ["topic_id", "round_id", "cord_uid", "relevance_score"]

    # Remove non necesary column (round_id)
    relevance_data = relevance_data.drop(["round_id"], axis=1)

    # Remove badly labeled examples (relevancy == -1)
    relevance_data = relevance_data[relevance_data.relevance_score >= 0]

    # Convert relevance score into binary
    relevance_data.loc[relevance_data["relevance_score"] == 2, "relevance_score"] = 1

    return relevance_data
```

# First Approach

As a first approach on the task of building and IR system, we are going to implement the well known **SVM** model. This is quite a simple model, but it can be useful for us in order to have a first approach and also, because it is one of the most widely used, as it is simple and provides good results.

So, the first we have to do to implement this model is to create a dictionary with the words that appear in the corpus. The dictionary is like a bag of unique words that appear y the documents.

In our case, we have create the corpus using the titles of the documents, and the abstract when there was no title, in order to reduce the high dimensionality of the problem. This are read from the previous documents generated before, containing the texts preprocessed.

Also, we want to note that when creating the dictionary, we've limited it in 20000 words. The decision to reduce to 20000 the number of terms of all the documents has been taken because computationaly it was very expensive to run the model. Specifically, creation of the Similarity Matrix and the computation of the Cosine similarity between all the titles.

```python
def load_document(path, field):
    doc = json.load(open(path))
    return doc[field]


def create_dictionary(docs):
    pdocs = [load_document("data/corpus/"+str(doc), 'title') for doc in docs]
    dictionary = corpora.Dictionary(pdocs, prune_at=20000)
    dictionary.save('/tmp/vsm.dict')
    return dictionary
```

After that, we now provide functions to create de Bag-of-Words, with the previous dictinary and we also create the TF-IDF model with that Bag-of-Words

```python
def docs2bows(corpus, dictionary):
    docs = [load_document("data/corpus/"+str(d), 'title') for d in corpus]
    vectors = [dictionary.doc2bow(doc) for doc in docs]
    corpora.MmCorpus.serialize('/tmp/vsm_docs.mm', vectors)
    return vectors


def create_TF_IDF_model(corpus):
    dictionary = create_dictionary(corpus) #corpora.MmCorpus.load('/tmp/vsm.dict')
    docs2bows(corpus, dictionary)
    loaded_corpus = corpora.MmCorpus('/tmp/vsm_docs.mm')
    tfidf = models.TfidfModel(loaded_corpus)
    return tfidf, dictionary
```

Finaly, we create two functions for evaluating the model. The first one, ranks the documents with a given quary and the second one makes all the evaluation process with all the querys we have.

```python
def launch_query(corpus, q, filename='/tmp/vsm_docs.mm'):
    tfidf, dictionary = create_TF_IDF_model(corpus)
    loaded_corpus = corpora.MmCorpus(filename)
    index = similarities.MatrixSimilarity(loaded_corpus, num_features=len(dictionary))
    vq = dictionary.doc2bow(q)
    qtfidf = tfidf[vq]
    sim = index[qtfidf]
    ranking = sorted(enumerate(sim), key=itemgetter(1), reverse=True)
    return ranking


def compute_real_pred_score(corpus, querys, relevance_judgements):
    for i in range(1, 51):
        ranking = launch_query(corpus, querys[i-1])
        indx = np.where(relevance_judgements['topic_id'].tolist()==np.int64(i))[0]
        topic_docs = np.array(relevance_judgements['cord_uid'])[indx]
        topic_real_score = np.array(relevance_judgements['relevance_score'])[indx]
        document_id = []
        scores = []
        data_equivalences = json.load(open('./data/cord_uid_sha_equivalences.json'))
        for rank in ranking:
            document_id.append(corpus[int(rank[0])].split('.')[0])
            scores.append(rank[1])
        dict_scores = dict()
        real_output = []
        prediction = []
        for num, doc in enumerate(topic_docs):
            if doc in data_equivalences.keys():
                j=0
                for documento in document_id:
                    if documento == data_equivalences[doc]:
                        break
                    j+=1
                score = np.array(scores)[j]
                real_output.append(str(topic_real_score[num]))
                prediction.append(str(round(score,3)))
        dict_scores['real_output'] = real_output
        dict_scores['prediction'] = prediction
        json.dump(dict_scores, open('./results_tfidf/'+'scores_topic_'+str(i)+'.json', 'w+'))
```

Now, we evaluate the model using the functions provided above, and save those results.

```python
# For computing the real score and the predicted score, execute it only once
compute_real_pred_score(corpus, querys, relevance_judgements)
```

Finaly, we provide two functions that read the computed results and shows the obtained results. We are showing here different metrics: The precision, recall, f1-score and the accuracy. That we think they can be the more representative ones.

```python
def evaluate_pred_topic(topic_id, json_path, th=0.05):
    results_path = './'+str(json_path)+'/scores_topic_'+str(topic_id)+'.json'
    data = json.load(open(results_path))
    y_true = list(map(int, data['real_output']))
    y_pred = []
    for pred in data['prediction']:
        if float(pred) > th:
            pred=1
        else:
            pred=0
        y_pred.append(pred)
    report = classification_report(y_true, y_pred, output_dict=True)
    return report

def display_results(json_path, type_model, thres=0.05):
    total_report = {'0': {'precision': 0.0, 'recall': 0.0, 'f1-score': 0.0, 'support': 0}, '1': {'precision': 0.0, 'recall': 0.0, 'f1-score': 0.0, 'support': 0}, 'accuracy': 0.0, 'macro avg': {'precision': 0.0, 'recall': 0.0, 'f1-score': 0.0, 'support': 0}, 'weighted avg': {'precision': 0.0, 'recall': 0.0, 'f1-score': 0.0, 'support': 0}}
    report = None
    for i in range(1,51):
        if type_model == 'vsm':
            report = evaluate_pred_topic(i, json_path, th=thres)
        elif type_model == 'kdtree':
            report = evaluate_pred_topic_kdtree(i, json_path)
        # elif type_model == 'word2vec':
        #     report = evaluate_pred_topic(i, json_path)

        for key in report['0'].keys():
            total_report['0'][key] += report['0'][key]
        for key in report['1'].keys():
            total_report['1'][key] += report['1'][key]
        total_report['accuracy'] += report['accuracy']
        for key in report['macro avg'].keys():
            total_report['macro avg'][key] += report['macro avg'][key]
        for key in report['weighted avg'].keys():
            total_report['weighted avg'][key] += report['weighted avg'][key]

    for key in report['0'].keys():
        if key != 'support':
            total_report['0'][key] = total_report['0'][key] / 50.0
    for key in report['1'].keys():
        if key != 'support':
            total_report['1'][key] = total_report['1'][key] / 50.0
    total_report['accuracy'] = total_report['accuracy'] / 50.0
    for key in report['macro avg'].keys():
        if key != 'support':
            total_report['macro avg'][key] = total_report['macro avg'][key] / 50.0
    for key in report['weighted avg'].keys():
        if key != 'support':
            total_report['weighted avg'][key] = total_report['weighted avg'][key] / 50.0
    print('########################################')
    print('  FINAL EVALUATION TOTAL RESULTS          ')
    print('  Results for class irrelevant:      ')
    for key in report['0'].keys():
        print('      '+ key+ ': ' +str(round(total_report['0'][key], 3))+'          ')
    print('  Results for class relevant:        ')
    for key in report['1'].keys():
        print('      '+ key+ ': ' +str(round(total_report['1'][key], 3))+'          ')
    print('  Results for Accuracy:              ')
    print('      Accuracy: ' +str(round(total_report['accuracy'], 3))+'          ')
    print('  Average results for all the topic ids:             ')
    for key in report['macro avg'].keys():
        print('      '+ key+ ': ' +str(round(total_report['macro avg'][key], 3))+'          ')
    print('  Results for weighted avg:          ')
    for key in report['weighted avg'].keys():
        print('      '+ key+ ': ' +str(round(total_report['weighted avg'][key], 3))+'          ')
    print('########################################')
```

# Second approach: KDTrees

In this second approach we decide to try an instance based model which name is KDTrees. This model is a variation of the classical KNN.

In this case, as this model is based on instances we decided to reduce the number of documetns in the corpus. Another important reason for us to take this decisions is that we note that in the relevance judgements provided, they evaluate less docuements than those that are in the original corpus, so we don't have information from many of them and we can not evaluate them.

So, the functions provided bellow are prepared to return only the papers that appear in the relevance judgements.

```python
def getIndexes(dfObj, value):
    ''' Get index positions of value in dataframe i.e. dfObj.'''
    listOfPos = list()
    # Get bool dataframe with True at positions where the given value exists
    result = dfObj.isin([value])
    # Get list of columns that contains the value
    seriesObj = result.any()
    columnNames = list(seriesObj[seriesObj == True].index)
    # Iterate over list of columns and fetch the rows indexes where value exists
    for col in columnNames:
        rows = list(result[col][result[col] == True].index)
        for row in rows:
            listOfPos.append((row, col))
    # Return a list of tuples indicating the positions of value in the dataframe
    return listOfPos


def extract_title(path_pdfs):
    titles = []
    for pdf in path_pdfs:
        pdf = './data/corpus/'+pdf+'.json'
        data = json.load(open(pdf))
        title = data['title']
        space = ' '
        if title:
            titles.append(space.join(title))
        else:
            titles.append(space.join(data['abstract']))
    return titles


def get_paper_id(data_equivalences, cord_uid):
    try:
        return data_equivalences[cord_uid]
    except KeyError as e:
        return None


def get_articles_per_topic(topic_id, relevance_judgements):
    indx = np.where(relevance_judgements['topic_id'].tolist()==np.int64(topic_id))[0]
    topic_docs = np.array(relevance_judgements['cord_uid'])[indx]
    return topic_docs


def get_article_list(relevance_judgements):
    articles = dict()
    for i in range(1, 51):
        articles[i] = get_articles_per_topic(i, relevance_judgements)
    return articles
```

And now we obtain those articles

```python
data_equivalences = json.load(open('./data/cord_uid_sha_equivalences.json'))
articles = get_article_list(relevance_judgements)
titles_topic = dict()
articles_id = []
relevance_score = []
for topic_id in range(1, 51):
    paper_id_list = list()
    for art_id in articles[topic_id]:
        paper_id = get_paper_id(data_equivalences, art_id)
        if paper_id:
            paper_id_list.append(paper_id)
            articles_id.append(art_id)
            indx = getIndexes(relevance_judgements, art_id)[0][0]
            relevance_score.append(relevance_judgements['relevance_score'][indx])
    titles_topic[topic_id] = extract_title(paper_id_list)
```

Once the selection of articles is done, we now train our KDTree model. After many tries, we note that even we reduce the number of documents, the model can not afford that corpus size (as we said this model is an instance based, so the time needed to train a big corpus can be unaffordable), we have to test the model with a little part of it. We have seen our shelf forced to do so, so we have reduced the feature size to 5000 terms in order to train correctly the model.
Here we provide the functions to train the model:

```python
def train_KDTree(titles_list, articles_id, relevance_score):
    #Counting the frequency of occurance of each word
    count_vector = CountVectorizer()
    df = pd.DataFrame (titles_list,columns=['title'])[:5000]

    train_counts = count_vector.fit_transform(df.title)

    #Using tf-idf to reduce the weight of common words
    tfidf_transform = TfidfTransformer()
    train_tfidf = tfidf_transform.fit_transform(train_counts)
    a = np.array(train_tfidf.toarray())

    #obtaining the KDTree
    kdtree = KDTree(a ,leaf_size=5)
    df['tfidf'] = list(train_tfidf.toarray())[:5000]
    df['cord_uid'] = articles_id[:5000]
    df['relevance_score'] = relevance_score[:5000]
    return kdtree, df


def train_corpus(titles_topic, articles_id, relevance_score):
    aux = list(titles_topic.values())
    aux2 = list(itertools.chain.from_iterable(aux))
    kdtree, df = train_KDTree(aux2, articles_id, relevance_score)
    return kdtree, df
```

And we make the training:

```python
kdtree, df = train_corpus(titles_topic, articles_id, relevance_score)
```

Finaly, as done before, we provide two functions to evaluate with all the quarys and to save those results:

```python
def evaluate_pred_topic_kdtree(topic_id, json_path):
    results_path = './'+str(json_path)+'/scores_topic_'+str(topic_id)+'.json'
    data = json.load(open(results_path))
    y_true = list(map(int, data['real_output']))
    y_pred = []
    for pred in data['prediction']:
        if float(pred) < 0.5:
            pred=1
        else:
            pred=0
        y_pred.append(pred)
    report = classification_report(y_true, y_pred, output_dict=True)
    return report

def evaluate_kdtree(df, kdtree, query):
    y_true = []
    y_pred = []
    for token in query:
        occurrences = np.where(df['title'].str.contains(token))[0]
        if occurrences.size > 0:
            distance, idx = kdtree.query(df['tfidf'][occurrences].tolist(), k=5)
            for i, value in list(enumerate(idx[0])):
                y_true.append(str(df['relevance_score'][value]))
                y_pred.append(str(distance[0][i]))
    return y_true, y_pred
```

```python
if not os.path.isdir("./results_kdtree"):
    os.makedirs('./results_kdtree')
```

And now we evaluate and print the results:

```python
# Execute it only once
for topic_id in range(50):
    dict_scores = dict()
    y_true, y_pred = evaluate_kdtree(df, kdtree, querys[topic_id])
    dict_scores['real_output'] = y_true
    dict_scores['prediction'] = y_pred
    json.dump(dict_scores, open('./results_kdtree/'+'scores_topic_'+str(topic_id + 1)+'.json', 'w+'))
```

# Third approach: Long Short-Term Memory deep neural network
The LSTM based models are widely used for NLP tasks. In this case, we provide here a model using an LSTM layer. We are going to try an LSTM network that is called Long Short-Term Memory, the network is going to be trained using supervised dataset and also tested on the trained set.

First step is to create the dataset for the training and the testing process, the real output is going to be a vector of length the number of classes and will have a number 1 or 0 in the corresponding index position that represents to the topic id. If the document is relevant on a topic_id, then the value of that index will be 1, otherwise will be 0

```python
def create_dataset(relevance_judgements):
    X = []
    y = []
    cord_uid = list(relevance_judgements['cord_uid'])
    topic_id = list(relevance_judgements['topic_id'])
    topic_real_score = list(relevance_judgements['relevance_score'])
    equivalences = json.load(open('./data/cord_uid_sha_equivalences.json'))
    for num, cord in enumerate(cord_uid):
        if cord in equivalences.keys():
            path = './data/corpus/'+equivalences[cord]+'.json'
            doc_title = json.load(open(path))['title']
            doc_title = ' '.join(doc_title)
            doc_abstract = json.load(open(path))['abstract']
            doc_abstract = ' '.join(doc_abstract)
            if doc_title != '':
                X.append(doc_title)
                y_aux = np.zeros((1, 50)).astype(int)[0]
                y_aux[topic_id[num] - 1] = topic_real_score[num]
                y.append(y_aux.tolist())
            elif doc_abstract != '':
                X.append(doc_abstract)
                y_aux = np.zeros((1, 50)).astype(int)[0]
                y_aux[topic_id[num] - 1] = topic_real_score[num]
                y.append(y_aux)
    return X, y
```

## Embedding Layer
First step is to create the tokenizer model to create a word-to-index dictionary where each word is used as a key and a unique index is used as the value for the key.

Second step is to use the GloVe embeddings file to create the feature matrix:
- File accessible from Kaggle: https://www.kaggle.com/danielwillgeorge/glove6b100dtxt

By loading the GloVe embeddings file we will create a dictinary with words as keys and their corresponding embedding list as values.

Finally, we will create an embedding matrix where each row number will correspond to the index of the word in the corpus. The matrix will have 100 columns where each column will contain the GloVe word embeddings for the words in our corpus. The number of columns can be changed depending on what we want but for this purpose it is correct to set it to 100, if one row has less than 100 columns the missing ones will be filled with zeros.

The embedding matrix will be the input weigts for training the models so it is very important to build it correctly

**Note:** The GloVe file must be placed in the same directory in which it is this notebook for the functions to work, and it is necessary to download it.

```python
MAX_LEN = 100
checkpoint_filepath_lstm = 'checkpoint/weights_lstm'

def create_checkpoint(path):
    model_checkpoint = ModelCheckpoint(
        filepath = path,
        save_weights_only = True,
        monitor = 'val_acc',
        mode = 'max',
        save_best_only = True
    )
    return model_checkpoint
    
def prepare_embedding_layer(X_train):
    tokenizer = Tokenizer(num_words=5000)
    tokenizer.fit_on_texts(X_train)
    X_train = tokenizer.texts_to_sequences(X_train)

    vocab_size = len(tokenizer.word_index) + 1 # unique words of the corpus
    X_train = pad_sequences(X_train, padding='post', maxlen=MAX_LEN)

    embeddings_dictionary = dict()
    glove_file = open('glove.6B.100d.txt', encoding="utf8")

    for line in glove_file:
        records = line.split()
        word = records[0]
        vector_dimensions = np.asarray(records[1:], dtype='float32')
        embeddings_dictionary[word] = vector_dimensions
    glove_file.close()

    embedding_matrix = np.zeros((vocab_size, 100))
    for word, index in tokenizer.word_index.items():
        embedding_vector = embeddings_dictionary.get(word)
        if embedding_vector is not None:
            embedding_matrix[index] = embedding_vector

    return X_train, embedding_matrix, vocab_size, tokenizer
```

```python
X, embedding_matrix, vocab_size, tokenizer = prepare_embedding_layer(X)
y = np.array(y)
```

## Printing graphs

1. Accuracy: comparison between the accuracy in validation and test
2. Loss: comparison between the loss in validation and test

```python
def print_graphs(out_values):
    plt.plot(out_values.history['acc'])
    plt.plot(out_values.history['val_acc'])

    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train','test'], loc = 'upper left')
    plt.show()

    plt.plot(out_values.history['loss'])
    plt.plot(out_values.history['val_loss'])

    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train','test'], loc = 'upper left')
    plt.show()
```

## Performance Metrics

This function takes as input the predictions and the real outputs and calculates returns some metric values to measure the performace of the model. This metrics are: TPD, FRP and AUC to calculate the ROC curve; as well as the confusion matrix.

```python
def get_performance_metrics(y_true, y_pred, n_classes):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_pred[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    return fpr, tpr, roc_auc
```

## Training the LSTM
Training and testing process has been done with the same sample
Train and test = X, y

```python
def create_model_lstm(vocab_size, embedding_matrix):
    model = Sequential()
    embedding_layer = Embedding(vocab_size, 100, weights=[embedding_matrix], input_length=MAX_LEN , trainable=False)
    model.add(embedding_layer)
    model.add(LSTM(128))
    # Depending on how the dataset has been built
    model.add(Dense(50, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

    return model
```

```python
def train_lstm(X_train, y_train, X_test, y_test, model):
    model_checkpoint_callback_lstm = create_checkpoint(checkpoint_filepath_lstm)
    out_values = model.fit(X_train, y_train, batch_size=128, epochs=6,
                           callbacks=[model_checkpoint_callback_lstm],
                           verbose=1, validation_split=0.2)

    score_test = model.evaluate(X_test, y_test, verbose=1)
    print("Test Score in the Recurrent Neural Network:", score_test[0])
    print("Test Accuracy in the Recurrent Neural Network:", score_test[1])
    print_graphs(out_values)

    return model
```

# Final Approach: Word2Vec model

Finally, we are again going to try with an VSM. More sepecifucally we are using the word2vec model. This is an state-of-the-art word emmbedding methos that has become very popular.

We can find an implementation of this model in the Gensim library, and that is what we are going to use in our case. So the first thing is to import the model:

```python
from gensim.models import Word2Vec
```

The first thing we have to do is prepare our data. We are going to use as an imput the same than with the LSTM: The documents from the relevance judgements that have either a title or an abstract (this is the information we decideto use before, due to computational simplicity).

The function provided before, prepares this input for word2vec model. So, as an input what we have is a list with the titles, each title is represented as a list os words.

```python
def prepare_data_w2v(X):
    train_set = list()
    for title in X:
        phrase = list()
        for token in title.split():
            phrase.append(token)
        train_set.append(phrase)
    return train_set
```

What we need to do now, is to obtain the vector representaton for each word in the corpus and from the querys. To do so, we provide the next function:

```python
# Function returning vector reperesentation of a document
def get_embedding_w2v(doc_tokens):
    embeddings = []
    if len(doc_tokens)<1:
        return np.zeros(300)
    else:
        for tok in doc_tokens:
            if tok in w2v_model.wv.vocab:
                embeddings.append(w2v_model.wv.word_vec(tok))
            else:
                embeddings.append(np.random.rand(300))
        # mean the vectors of individual words to get the vector of the document
        return np.mean(embeddings, axis=0)
```

The next step is to provide functions that can measure the similarity between a given query and the documents in the corpus. To do so, we are using the cosine similarity:

```python
from sklearn.metrics.pairwise import cosine_similarity
def calculate_similarity(corpus, query):    
    results = list()
    for title in corpus:
        res = cosine_similarity(title[0].reshape(1,-1), query.reshape(1,-1)).item()
        results.append(res)

    return np.array(results)
```

```python
def compute_real_pred_score(corpus, querys, relevances):
    for i in range(1, 51):
        print ('Query: ', i)
        indx = np.where(relevances['topic_id'].tolist()==np.int64(i))[0]
        ranking = calculate_similarity(corpus, querys[i-1][0])[indx]
#         print (ranking)
        topic_real_score = np.array(relevances['relevance_score'])[indx]
        dict_scores = dict()
        dict_scores['real_output'] = list(map(str, topic_real_score))
        dict_scores['prediction'] = list(map(str, ranking))
#         print (dict_scores)
        json.dump(dict_scores, open('./results_word2vec/'+'scores_topic_'+str(i)+'.json', 'w+'))
```
